```python
from bson import ObjectId
```

```python
data = []
with open(file_path, "r", encoding='utf-8') as file:
    for line in file:
      record = json.loads(line)
# Converter o campo '_id' com o $oid em ObjectId
      if '_id' in record and '$oid' in record['_id']:
        record['_id'] = ObjectId(record['_id']['$oid'])
      data.append(record)
```

```python
collection.insert_many(data)
print("Dados inseridos com sucesso!")
```

```python
#Exercício 20
restaurante = collection.find({"name": "333 Chinese Takeaway"})

ids_para_remover = []
ids_para_manter = set()

for doc in restaurante:
    if doc['_id'] not in ids_para_manter:
        ids_para_manter.add(doc['_id'])  # Manter o primeiro encontrado
    else:
        ids_para_remover.append(doc['_id'])  # Adicionar à lista de remoção

collection.delete_many({"_id": {"$in": ids_para_remover}})
```
